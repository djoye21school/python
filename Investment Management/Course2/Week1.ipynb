```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import edhec_risk_kit_204 as erk

%load_ext autoreload
%autoreload 2
ind49_rets = erk.get_ind_returns(weighting="vw", n_inds=49)["1991":"2018"]
fff = erk.get_fff_returns()["1991":"2018"]
beer_excess = ind49_rets.loc[:,["Beer"]] - fff.loc[:, ['RF']].values   
mkt_excess = fff.loc[:,['Mkt-RF']]
exp_var = mkt_excess.copy()
exp_var["Constant"] = 1
lm = sm.OLS(beer_excess, exp_var).fit()
lm.summary()
```

```python
ind49_rets = erk.get_ind_returns(weighting="vw", n_inds=49)["1991":"1993"]
cl=list(ind49_rets.columns)

Betas={'Industry':cl,'Beta':np.repeat(0,len(cl))}
Beta=pd.DataFrame(data=Betas)

mkt_excess = fff.loc["1991":"1993",['Mkt-RF']]
exp_var = mkt_excess.copy()
exp_var["Constant"] = 1
count=0
n=len(cl)

for i in range(0,n):
    index_excess = ind49_rets.loc["1991":"1993",[f"{cl[i]}"]].values - fff.loc["1991":"1993", ['RF']].values
    lm5 = sm.OLS(index_excess, exp_var).fit()
    Beta.loc[count,"Beta"]=lm5.params[0]
    count+=1
    
```

```python
ind49_rets = erk.get_ind_returns(weighting="vw", n_inds=49)["1991":"2018"]
fff = erk.get_fff_returns()["1991":"2018"]
excess = ind49_rets - fff.loc[:, ['RF']].values
rf = fff.loc[:, ['RF']].values
mkt_excess = fff.loc[:,['Mkt-RF']]
exp_var = mkt_excess.copy()
exp_var["Constant"] = 1
alloc = pd.DataFrame(index = ["beta"], columns =ind49_rets.columns)
exp_var["Value"] = fff.loc["1991":"2018",['HML']]
exp_var["Size"] = fff.loc["1991":"2018",['SMB']]

for i in ind49_rets.columns:
    excess = ind49_rets.loc[:,[i]] - rf
    lm = sm.OLS(excess, exp_var).fit()
    alloc[i] = lm.params[3]
    
    
alloc.T[alloc.T == alloc.T.max()]  
```
